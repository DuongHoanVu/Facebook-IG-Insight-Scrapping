```python
post_url_dict = {'image_url':[],
                 'image_source':[],
                 'post_url':[],
                 'likes_num':[],
                 'comments_num':[],
                 'date':[]}

for each_post in list_posts:
    image_url = each_post.find_element_by_css_selector('div[class="mtm"]').find_element_by_tag_name('a').get_attribute('href').split('?')[0]
    image_url = image_url.replace('https://www.', '')
    if image_url[-1] == '/':
        image_url = image_url[:-1]
    
    image_source = each_post.find_element_by_css_selector('div[class="mtm"]').find_element_by_tag_name('img').get_attribute('src')
    
    post_url_clean = each_post.find_element_by_class_name('_5pcq').get_attribute('href').split('?')[0]
    post_url_clean = post_url_clean.replace('https://www.', '')
    if post_url_clean[-1] == '/':
        post_url_clean = post_url_clean[:-1]
    
    try:
        likes_num = each_post.find_element_by_class_name('_7a9u').find_element_by_class_name('_66lg').find_element_by_css_selector('span[class="_81hb"]').text
    except:
        likes_num = 0

    try:
        comments_num =each_post.find_element_by_class_name('_7a9u').find_element_by_class_name('_4vn1').find_element_by_css_selector('span[class="_4vn2"]').text
        comments_num = comments_num.replace(' Comments', '')
    except:
        comments_num = 0
    
    date = each_post.find_element_by_tag_name('abbr').get_attribute('title')
    date = date.split(',', 1)[1]
    date = pd.to_datetime(date)
    
    post_url_dict['image_url'].append(image_url)
    post_url_dict['image_source'].append(image_source)
    post_url_dict['post_url'].append(post_url_clean)
    post_url_dict['likes_num'].append(likes_num)
    post_url_dict['comments_num'].append(comments_num)
    
    post_url_dict['date'].append(date)
```
